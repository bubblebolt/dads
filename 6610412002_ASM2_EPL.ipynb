### Step 1. Import the necessary libraries**

```python
import pandas as pd
import numpy as np


epl = pd.read_csv("https://raw.githubusercontent.com/bubblebolt/dads/main/DADS5001/2021-2022.csv")
epl
```

# Question 1
คำถาม: จำนวนประตูทั้งหมดที่ทำได้ใน EPL ฤดูกาล 2021-2022 เป็นเท่าใด

```python
#คำนวณจำนวนประตูทั้งหมด
total_goals = epl['FTHG'].sum() + epl['FTAG'].sum()

print("Home Team goals scored in the EPL 2021-2022 :", epl['FTHG'].sum())
print("Away Team goals scored in the EPL 2021-2022 :", epl['FTAG'].sum())

print("Total goals scored in the EPL 2021-2022 :", total_goals)
```

```python
#แสดงผลเป็น Bar Chart
import plotly.express as px

data1 = {
    'Team': ['Home Team', 'Away Team', 'Total Goals'],
    'Goals': [epl['FTHG'].sum(), epl['FTAG'].sum(), total_goals]
}

df_goals = pd.DataFrame(data1)

fig = px.bar(df_goals, x='Team', y='Goals', text='Goals',color='Goals', title='Goals Scored in EPL 2021-2022',
             labels={'Goals': 'Number of Goals'})

fig.show()
```

# Question 2:
คำถาม: ในฤดูกาล EPL 2021-2022 มีกี่นัดเตะที่จบลงด้วยผลเสมอ?


```python
#คำนวณผลแพ้ชนะในฤดูกาล EPL 2021-2022
home_win = epl[epl['FTR'] == 'H']['FTR'].count()
away_win = epl[epl['FTR'] == 'A']['FTR'].count()
draw_matches = epl[epl['FTR'] == 'D']['FTR'].count()

print("จำนวนนัดเตะที่จบลงด้วยผลทีมเจ้าบ้านชนะใน EPL ฤดูกาล 2021-2022:", home_win)
print("จำนวนนัดเตะที่จบลงด้วยผลทีมเยือนชนะใน EPL ฤดูกาล 2021-2022:", away_win)
print("จำนวนนัดเตะที่จบลงด้วยผลเสมอใน EPL ฤดูกาล 2021-2022:", draw_matches)
```

```python
#แสดงผลเป็น Pie Chart
data2 = {
    'Result': ['Home Win', 'Away Win', 'Draw'],
    'Count': [home_win, away_win, draw_matches]
}

df_results = pd.DataFrame(data2)

fig = px.pie(df_results, values='Count', names='Result',
             title='Results of EPL Matches 2021-2022',
             labels={'Count': 'Number of Matches'})

fig.show()
```

# Question 3:
คำถาม: ทีมใดมีผลต่างประตูได้เสียสูง 5 อันดับแรกใน EPL ฤดูกาล 2021-2022?

```python
#คำนวณผลต่างได้เสียแต่ละทีม
epl['HGoalFor'] = epl['FTHG']
epl['HGoalAgainst'] = epl['FTAG']
epl['AGoalFor'] = epl['FTAG']
epl['AGoalAgainst'] = epl['FTHG']

# คำนวณคะแนนสองคอลัมเป็นคะแนนประตูที่ได้ (Goal For) และคะแนนประตูที่เสีย (Goal Against)
team_goal_difference_home = epl.groupby('HomeTeam').agg({'HGoalFor': 'sum', 'HGoalAgainst': 'sum'}).reset_index()
team_goal_difference_home.columns = ['Team', 'GoalFor', 'GoalAgainst']
team_goal_difference_away = epl.groupby('AwayTeam').agg({'AGoalFor': 'sum', 'AGoalAgainst': 'sum'}).reset_index()
team_goal_difference_away.columns = ['Team', 'GoalFor', 'GoalAgainst']

result_df = pd.concat([team_goal_difference_home, team_goal_difference_away]).groupby('Team').agg({'GoalFor': 'sum', 'GoalAgainst': 'sum'}).reset_index()

result_df['Difference'] = result_df['GoalFor'] - result_df['GoalAgainst']

result5 = result_df.sort_values(by='Difference', ascending=False).head(5)
```

```python
#แสดงผลเป็น Bar chart แบบกลุ่ม
fig = px.bar(result5, x='Team', y=['GoalFor', 'GoalAgainst', 'Difference'],
             labels={'value': 'Goals'},
             title='Goal Difference of Top 5 Teams in EPL 2021-2022',
             barmode='group')

fig.show()
```

# Question 4:
คำถาม : ตัวแปรต้นใดมีผลต่อผลลัพธ์มากที่สุดในการทำนายผลการแข่งขันในฤดูกาล EPL 2021-2022?
ใช้เทคนิคทางสถิติ (เช่น correlation coefficient) เพื่อวัดความสัมพันธ์ระหว่างตัวแปรต้นและผลลัพธ์. ตัวแปรที่มีค่า correlation coefficient สูงที่สุดจะถือว่ามีผลต่อผลลัพธ์มากที่สุด.

```python
# คำนวณ correlation coefficient
selected_columns = ['FTHG', 'FTAG', 'HS', 'AS', 'HST', 'AST', 'HC', 'AC', 'HF', 'AF', 'HY', 'AY', 'HR']
epl_selected = epl[selected_columns]

correlation_matrix = epl_selected.corr()
correlation_all = correlation_matrix.unstack().sort_values(ascending=False)

# เลือกตัวแปรที่เกี่ยวข้องกับผลลัพธ์ (FTHG และ FTAG) และแสดงผลลัพธ์
correlation_FTHG = correlation_all.loc[('FTHG',)]
correlation_FTAG = correlation_all.loc[('FTAG',)]

correlation_FTHG, correlation_FTAG
```

 **จะเห็นได้ว่า**


*   สำหรับ FTHG (Full Time Home Team Goals): ตัวแปรที่มี correlation coefficient บวกมากที่สุดคือ HST (Home Team Shots on Target) ที่มีค่า 0.636086
ตัวแปรที่มี correlation coefficient ลบมากที่สุดคือ AS (Away Team Shots) ที่มีค่า -0.279169
*   สำหรับ FTAG (Full Time Away Team Goals):
ตัวแปรที่มี correlation coefficient บวกมากที่สุดคือ AST (Away Team Shots on Target) ที่มีค่า 0.650264
ตัวแปรที่มี correlation coefficient ลบมากที่สุดคือ HS (Home Team Shots) ที่มีค่า -0.285164






```python
#แสดงผลัพธ์ในรูบแปบบ heatmap
import seaborn as sns
import matplotlib.pyplot as plt

correlation_matrix = epl_selected.corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5, square=True)
plt.title('Correlation Heatmap of Selected Variables')
plt.show()
```

# Question 5:
คำถาม: จากผลการแข่งขันและภาพรวมของทีมในฤดูกาล EPL 2021-2022 สร้างโมเดลเพื่อทำนายผลการแข่งขันในฤดูกาลถัดไป
พิจารณาที่จะใช้ Multiple Linear Regression เพื่อทำนายผลการแข่งขันทีมในฤดูกาลถัดไปใน English Premier League (EPL)

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics

# สร้าง DataFrame ที่มีเฉพาะตัวแปรต้นที่สนใจ (HST, HS, HC, AST, AS)
X = epl[['HST', 'HS', 'HC', 'AST', 'AS']]

# กำหนดตัวแปรตาม (FTHG, FTAG)
y_FTHG = epl['FTHG']
y_FTAG = epl['FTAG']

X_FTHG_train, X_FTHG_test, y_FTHG_train, y_FTHG_test = train_test_split(X, y_FTHG, test_size=0.2, random_state=0)
X_FTAG_train, X_FTAG_test, y_FTAG_train, y_FTAG_test = train_test_split(X, y_FTAG, test_size=0.2, random_state=0)

# สร้างและฝึก Linear Regression Model
model_FTHG = LinearRegression()
model_FTHG.fit(X_FTHG_train, y_FTHG_train)

model_FTAG = LinearRegression()
model_FTAG.fit(X_FTAG_train, y_FTAG_train)


y_FTHG_pred = model_FTHG.predict(X_FTHG_test)
y_FTAG_pred = model_FTAG.predict(X_FTAG_test)


print('FTHG Mean Absolute Error:', metrics.mean_absolute_error(y_FTHG_test, y_FTHG_pred))
print('FTAG Mean Absolute Error:', metrics.mean_absolute_error(y_FTAG_test, y_FTAG_pred))
print('Coefficients for FTHG:', model_FTHG.coef_)
print('Coefficients for FTAG:', model_FTAG.coef_)
```

```python
# สร้าง Scatter plot และ trend line
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.regplot(x=y_FTHG_test, y=y_FTHG_pred, scatter_kws={'alpha':0.3})
plt.title('Actual vs Predicted FTHG')
plt.xlabel('Actual FTHG')
plt.ylabel('Predicted FTHG')
plt.subplot(1, 2, 2)
sns.regplot(x=y_FTAG_test, y=y_FTAG_pred, scatter_kws={'alpha':0.3})
plt.title('Actual vs Predicted FTAG')
plt.xlabel('Actual FTAG')
plt.ylabel('Predicted FTAG')

# แสดงสมการ Linear Regression
print("FTHG Regression Equation:")
print(f"FTHG = {model_FTHG.intercept_:.2f} + {model_FTHG.coef_[0]:.2f} * HST + {model_FTHG.coef_[1]:.2f} * HS + {model_FTHG.coef_[2]:.2f} * HC + {model_FTHG.coef_[3]:.2f} * AST + {model_FTHG.coef_[4]:.2f} * AS")

print("\nFTAG Regression Equation:")
print(f"FTAG = {model_FTAG.intercept_:.2f} + {model_FTAG.coef_[0]:.2f} * HST + {model_FTAG.coef_[1]:.2f} * HS + {model_FTAG.coef_[2]:.2f} * HC + {model_FTAG.coef_[3]:.2f} * AST + {model_FTAG.coef_[4]:.2f} * AS")

plt.show()
```
